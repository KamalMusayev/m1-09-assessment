## Clean pm25

#### Coerce invalid pm25 strings to NaN. Show the number of invalid values converted.

```python
print("Info: ")
print(df_copy.info())
print("======================")
print("Describe: ")
print(df_copy.describe())
print("======================")
print("Missing values count: ")
print(df_copy.isna().sum())
print("======================")
#I used before_conversion for showing the number of invalid values converted.
before_conversion = df["pm25"].isna().sum()
df_copy["pm25"] = pd.to_numeric(df_copy["pm25"], errors="coerce")

after_conversion = df_copy["pm25"].isna().sum()

converted_invalids = after_conversion - before_conversion

print("Invalid values converted to NaN:", converted_invalids)
```

## Handle missing values

#### Choose a missing-value strategy for pm25 and justify it in a short markdown cell. Apply it and confirm the new missing count.

```python
# TODO: handle missing values in pm25 and justify in markdown
df_copy["pm25"] = df_copy["pm25"].interpolate(method="time")
df_copy["pm25"].isna().sum()
```

### I used time-based interpolation to fill missing pm25 values because it is a time-dependent variable that changes gradually from day to day.

### This method preserves the natural trend in the data and avoids unrealistic assumptions like using a global average or assuming no change between consecutive days.

## Data quality analysis

#### Determine which city has the highest percentage of invalid or missing pm25 readings. Show the calculation.

```python
# I used another copy of df because we filled nan values for pm25 in the first copy of df.
df_copy2 = df.copy()

nan_percent_by_city = df_copy2.groupby("city")["pm25"].apply(lambda x: x.isna().mean() * 100)
print("Percentage of missing pm25 by city:")
print(nan_percent_by_city)

max_nan_city = nan_percent_by_city.idxmax()
max_nan_percent = nan_percent_by_city.max()

print(f"\nCity with the highest percentage of missing pm25: {max_nan_city} ({max_nan_percent:.2f}%)")
```

## Rolling time-series analysis

#### Compute a 7‑day rolling average of pm25 per city. Explain in 2–3 sentences why rolling averages help interpretation.

```python
"""I used `df_copy`, which is a copy of the original dataset where I had filled 
the NaN values. Computing a rolling average on data with missing values 
does not make sense, because any window containing NaNs would result in NaN, 
making trends harder to interpret."""
seven_day_of_average = (
    df_copy.groupby("city")["pm25"]
    .rolling(7)
    .mean()
)
seven_day_of_average[:15]
```

#### Rolling averages help us because they smooth out daily ups and downs in the data. This makes it easier to see the overall trend instead of focusing on small random changes. They help us understand the bigger pattern more clearly.

## Event detection

#### Define a "high pollution event" using a percentile‑based threshold of pm25. Count events per city and show the threshold you used.

```python
threshold = df_copy["pm25"].quantile(0.9)
counts_percentile = df_copy.groupby("city")["pm25"].apply(lambda x: (x > threshold).sum())
counts_percentile
```

## Volatility comparison

#### Choose two cities and decide which has more volatile air quality over time. Define your metric (e.g., rolling standard deviation) and justify your conclusion.

```python
"""I measured the volatility of PM2.5 in each city using the standard deviation. 
Cities with higher standard deviation have more day-to-day fluctuations in air pollution,
while lower values indicate more stable air quality. This gives an easy way to compare
how variable pollution is between cities."""
first_city_pm25 = df_copy[df_copy["city"] == "Berlin"]["pm25"]
second_city_pm25 = df_copy[df_copy["city"] == "London"]["pm25"]
first_city_std = first_city_pm25.std()
second_city_std = second_city_pm25.std()
if first_city_std > second_city_std:
    print(f"Berlin has more volatility ({first_city_std:.2f} vs {second_city_std:.2f})")
else:
    print(f"London has more volatility ({second_city_std:.2f} vs {first_city_std:.2f})")
```

## Line plot
#### Plot monthly pm25 trends for at least two cities.

```python
monthly_avg = df_copy.groupby(["city", "month"])["pm25"].mean().reset_index()

import matplotlib.pyplot as plt

cities = ["Berlin", "London", "Madrid"]

plt.figure()

for city in cities:
    city_data = monthly_avg[monthly_avg["city"] == city]
    plt.plot(city_data["month"], city_data["pm25"], label=city)

plt.xlabel("Month")
plt.ylabel("Average PM2.5")
plt.title("Monthly PM2.5 Trend")
plt.legend()
plt.show()
```

## Bar chart

#### Plot overall average pm25 by city.

```python
average_pm25_by_city = df_copy.groupby("city")["pm25"].mean()

plt.figure()
average_pm25_by_city.plot(kind="bar")
plt.xlabel("City")
plt.ylabel("Average PM2.5")
plt.title("Average PM2.5 by City")
plt.show()
```
